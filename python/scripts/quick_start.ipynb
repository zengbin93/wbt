## 绘图示例

使用 `wbt.plotting` 模块可视化回测结果。

```python
from wbt.plotting import (
    plot_backtest_overview,
    plot_colored_table,
    plot_cumulative_returns,
    plot_daily_return_dist,
    plot_drawdown,
    plot_long_short_comparison,
    plot_monthly_heatmap,
    plot_pairs_analysis,
    plot_symbol_returns,
)
```

```python
# 累计收益曲线
plot_cumulative_returns(wb.daily_return, cols=["total"])
```

```python
# 回撤分析
plot_drawdown(wb.daily_return, col="total")
```

```python
# 日收益分布
plot_daily_return_dist(wb.daily_return, col="total")
```

```python
# 月度收益热力图
plot_monthly_heatmap(wb.daily_return, col="total")
```

```python
# 品种收益分布
plot_symbol_returns(wb.dailys)
```

```python
# 交易分析
plot_pairs_analysis(wb.pairs)
```

```python
# 回测概览（2x2 子图：回撤+累计收益、日收益分布、月度热力图）
plot_backtest_overview(wb.daily_return)
```

```python
# 绩效指标表格（红涨绿跌）
plot_colored_table(wb.stats)
```

```python
# 多空对比（累计收益曲线 + 关键指标对比表）
plot_long_short_comparison(wb.daily_return, wb.stats, wb.long_stats, wb.short_stats)
```
