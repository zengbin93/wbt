# Weight Backtest 快速入门
---

核心功能：

1. 使用 `backtest` 函数执行回测，并查看结果
2. 使用 `plotting` 模块查看回测结果相关图表


```python
import wbt
```

```python
# 生成示例权重数据（自洽可跑）；也可替换为自己的 DataFrame
df = wbt.mock_weights(freq="30分钟")
df.head()
```

```python
wb = wbt.backtest(df)
```

```python
wb.stats
```

```python
wb.segment_stats(sdt="2020-01-01", edt="2020-12-31", kind="多空")
```

```python
wb.long_alpha_stats
```

## 绘图示例

所有绘图函数以 `BacktestResult` 为统一输入：`result = wb.to_result()`，下游零数据转换。

```python
from wbt.plotting import (
    plot_backtest_overview,
    plot_colored_table,
    plot_cumulative_returns,
    plot_daily_return_dist,
    plot_drawdown,
    plot_drawdowns_table,
    plot_key_trades,
    plot_long_short_comparison,
    plot_monthly_heatmap,
    plot_pairs_analysis,
    plot_symbol_returns,
    plot_verdict,
)

# 标准输入：所有绘图共用同一个 result
result = wb.to_result()
```

```python
# 累计收益曲线
plot_cumulative_returns(result, keys=["多空"])
```

```python
# 回撤分析
plot_drawdown(result, key="多空")
```

```python
# 日收益分布
plot_daily_return_dist(result)
```

```python
# 月度收益热力图
plot_monthly_heatmap(result)
```

```python
# 品种收益分布
plot_symbol_returns(result)
```

```python
# 交易分析
plot_pairs_analysis(result)
```

```python
# 回测概览（2x2 子图：回撤+累计收益、日收益分布、月度热力图）
plot_backtest_overview(result)
```

```python
# 绩效指标表格（红涨绿跌）
plot_colored_table(result)
```

```python
# 多空对比（原始 + 波动率归一曲线 + 关键指标对比表）
plot_long_short_comparison(result)
```

```python
# 关键交易：每年最赚/最亏各 3 笔
plot_key_trades(result)
```

```python
# 回撤明细表（审核页面）
plot_drawdowns_table(result)
```

```python
# 策略判定 + 年度指标（审核页面）
plot_verdict(result)
```
